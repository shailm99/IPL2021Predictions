```python
import pandas as pd
import numpy as np
```

```python
matches = pd.read_csv('IPL Matches 2008-2020.csv', index_col = 'id')
matches.head()
```

```python
import datetime 
matches.date = pd.to_datetime(matches.date)
matches = matches[matches.date > np.datetime64('2018-01-01')]
```

```python
delivery = pd.read_csv('IPL Ball-by-Ball 2008-2020.csv', index_col = 'id')
delivery.head()
delivery = delivery.loc[matches.index]
delivery.batting_team[delivery.batting_team == 'Delhi Daredevils'] = 'Delhi Capitals'
delivery.bowling_team[delivery.bowling_team == 'Delhi Daredevils'] = 'Delhi Capitals'
delivery
```

```python
matches.team1[matches.team1 == 'Delhi Daredevils'] = 'Delhi Capitals'
matches.team2[matches.team2 == 'Delhi Daredevils'] = 'Delhi Capitals'
matches.winner[matches.winner == 'Delhi Daredevils'] = 'Delhi Capitals'
matches.head()
```

```python
matches['Team1Won'] = matches.team1 == matches.winner
matches['Team2Won'] = matches.team2 == matches.winner
```

```python
game_lineups = list(zip(list(matches['team1']), list(matches['team2'])))
games = []
for game in game_lineups:
    games.append('_'.join(sorted(game)))

matches['team_team'] = games
```

```python
result = []
for i in range(len(matches)):
    result.append(list(matches['team_team'].str.split('_'))[i][0] == matches['winner'].iloc[i])
matches['result'] = result
matches
```

```python
overall_record = pd.DataFrame(columns = ['Games Won', 'Games Played'], index = np.arange(8))
overall_record.index = matches.groupby('team1').aggregate({'Team1Won' : 'sum'}).index
overall_record['Games Won'] = matches.groupby('team1').aggregate({'Team1Won' : 'sum'}).values.reshape(-1,1)
overall_record['Games Played'] = matches.groupby('team1').aggregate({'Team1Won' : 'count'}).values.reshape(-1,1)
overall_record
```

```python
overall_record['Games2 Won'] = matches.groupby('team2').aggregate({'Team2Won' : 'sum'}).values.reshape(-1,1)
overall_record['Games2 Played'] = matches.groupby('team2').aggregate({'Team2Won' : 'count'}).values.reshape(-1,1)
```

```python
overall_record['Won'] = overall_record['Games2 Won'] + overall_record['Games Won']
overall_record['Played'] = overall_record['Games2 Played'] + overall_record['Games Played']
overall_record.drop(overall_record.columns[:4], axis = 1, inplace = True)
overall_record
```

```python
overall_record['Percentage'] = overall_record['Won'] / overall_record['Played']
overall_record
```

```python
def find_player_stats(demo):
    play = demo.split()[-1]
    data = []
    if delivery[delivery.batsman.str.contains(play)].batsman.nunique() > 1:
        players = delivery[delivery.batsman.str.contains(play)].batsman.unique()
        first_letter = demo[0]
        reduced = []
        for plays in players:
            if first_letter == plays[0]:
                reduced.append(plays)
        if len(reduced) == 1:
            play = reduced[0]
        else:
            return False
    
    if delivery[(delivery.batsman.str.contains(play)) | (delivery.bowler.str.contains(play))].index.nunique() < 22:
        return False
    data.append(delivery[delivery.batsman.str.contains(play)].batsman_runs.sum()) # runs scored
    data.append(delivery[delivery.batsman.str.contains(play)].ball.count()) # balls faced
    data.append(delivery[delivery.batsman.str.contains(play)].is_wicket.sum()) # dismissed

    data.append(delivery[delivery.bowler.str.contains(play)].total_runs.sum()) # runs conceded
    data.append(delivery[delivery.bowler.str.contains(play)].ball.count()) # balls bowled
    data.append(delivery[delivery.bowler.str.contains(play)].is_wicket.sum()) # wickets taken

    data.append(True) # added values 
    return data
```
